### <p style="background-color:#FFFFFF; font-family:newtimeroman; color:blue; font-size:115%; text-align:left;"><strong>üëâ About Dataset:</strong></p>

#### Context:

**[The World Happiness Report](https://worldhappiness.report/about/#:~:text=They%20include%20GDP%20per%20capita,freedom%2C%20generosity%2C%20and%20corruption.)** is a landmark survey of the state of global happiness. The World Happiness Report is a project of the United Nations Sustainable Development Solutions Network (SDSN). It is an **``annual publication``** that aims to **``assess and compare the happiness levels of countries worldwide``**. The **``first report``** was published in **``2012``**, the second in ``2013``, the third in ``2015``, and the fourth in the ``2016`` Update. The World Happiness ``2017``, which ranks 155 countries by their happiness levels, was released at the United Nations at an event celebrating International Day of Happiness on March 20th.

The World Happiness Report attempts to **``measure the happiness level of countries by using various factors``**. These factors include **``income levels``**, **``social support``**, **``life expectancy``**, **``freedom``**, **``generosity``**, **``corruption levels``**, and **``social support``**. The report **``ranks countries using the "Happiness Score"``** derived from the combination of these indicators.

The purpose of the report is to encourage policymakers and community leaders to develop policies that promote happiness. It also aims to **``understand the relationship between happiness and social, economic, and environmental factors by identifying and comparing happiness levels across different countries``**.

The World Happiness Report is created by analyzing various data sources. Data obtained from surveys, databases, and other sources are used to determine the happiness levels of countries. The report provides a ranking among countries and **``highlights specific factors that influence happiness levels``**.

The World Happiness Report is considered an important resource to guide governments in the policy-making process and enhance the well-being of societies. It has become a widely recognized reference worldwide and contributes to the development of effective policies that impact happiness.

#### Content:

The happiness scores and rankings use data from the **[Gallup World Poll](https://www.gallup.com/analytics/318875/global-research.aspx)**. The columns following the happiness score estimate the extent to which each of six factors ‚Äì economic production, social support, life expectancy, freedom, absence of corruption, and generosity ‚Äì contribute to making life evaluations higher in each country than they are in Dystopia, a hypothetical country that has values equal to the world‚Äôs lowest national averages for each of the six factors. They have NO impact on the total score reported for each country, but they do explain why some countries rank higher than others.

#### What is Dystopia?

**[Dystopia](https://en.wikipedia.org/wiki/Dystopia)** is an **``imaginary country``** that has the **``world‚Äôs least-happy people``**. The purpose in establishing Dystopia is to have a benchmark against which all countries can be favorably compared (no country performs more poorly than Dystopia) in terms of each of the six key variables, thus allowing each sub-bar to be of positive width. **``The lowest scores observed for the six key variables, therefore, characterize Dystopia``**. Since **``life would be very unpleasant in a country with the world‚Äôs lowest incomes, lowest life expectancy, lowest generosity, most corruption, least freedom and least social support``**, it is referred to as **``‚ÄúDystopia,‚Äù``** in contrast to Utopia. Additionally, dystopia was used as an antonym for **[Utopia](https://en.wikipedia.org/wiki/Utopia)** by **[John Stuart Mill](https://en.wikipedia.org/wiki/John_Stuart_Mill)** in one of his 1868 Parliamentary Speeches (Hansard Commons) by adding the prefix "dys" (Ancient Greek: Œ¥œÖœÉ- "bad") to "topia"

- **``75th Percentile (Q3)``**: The value below which 75% of the data falls.
- **``Maximum``**: The maximum value in the column.

These statistics provide insights into the distribution of the data, including measures of **``central tendency``** (mean, median), **``spread``** (standard deviation), and **``outliers``** (minimum, maximum).

**Our function ``first_looking()`` is designed to provide a ``comprehensive overview`` of a Pandas DataFrame, including its ``shape``, ``null values``, ``unique values``, and ``value counts`` for each column.**

````python
def first_looking(df):
    """
    Prints basic information about a column in a Pandas DataFrame.
    Parameters:
    -----------
    df : pandas.DataFrame
        The DataFrame to be analyzed.
    """
    print(Fore.BLUE + "Shape of DataFrame", Fore.RESET)
    print(Fore.GREEN + "There have been " + Fore.RED + f"{df.shape[0]}" + Fore.GREEN + " rows and " + Fore.RED + f"{df.shape[1]}" + Fore.GREEN + " columns" + Fore.RESET)
    print("_"*40)
    print(Fore.BLUE + "Number of NULL values in the features\n", Fore.RESET)
    for i in df.columns:
        print(f"{i:<30}: {df[i].astype(str).isnull().sum()}")
    print("_"*40)
    print(Fore.BLUE + "Percent of NULL values in the features\n", Fore.RESET)
    for column in df.columns:
        print(f"{column:<30}: {df[column].isnull().mean()*100}")
    print("_"*40)
    print(Fore.BLUE + "Number of UNIQUE values in the features\n", Fore.RESET)
    for i in df.columns:
        print(f"{i:<30}: {df[i].astype(str).nunique()}")
    print("_"*40)
    print(Fore.BLUE + "Value Counts of Each Feature:\n", Fore.RESET)
    for i in df.columns:
        print(Fore.GREEN + f"column name: {Fore.RED}{i}{Fore.RESET}")
        value_counts_df = df[i].value_counts(dropna=False).reset_index()
        aligned_value_counts_str = "\n".join([f"{index_val:<35}: {count}" for index_val, count in zip(value_counts_df.iloc[:, 0], value_counts_df.iloc[:, 1])])
        print(aligned_value_counts_str)
        print("\n")
    print("_"*40)
````

- **``Explanation of Each Section``**
  - **``Shape of DataFrame:``**
    - Prints the number of rows and columns in the DataFrame.
  - **``Number of NULL Values``**:
    - Iterates through each column and prints the count of missing values.
  - **``Percent of NULL Values``**:
    - Calculates and prints the percentage of missing values for each column.
  - **``Number of UNIQUE Values``**:
    - Prints the count of unique values in each column.
  - **``Value Counts``**:
    - For each column, it prints the count of each unique value, including missing values (if any).

  - **``Calculate the Correlation Matrix``**: This gives us a matrix of correlation coefficients between pairs of features.
  - **``Identify the Two Features with the Highest Correlation``**: Extract the names of the two features that have the highest correlation value.
  - **``Create the New Feature``**: Compute the square root of the product of these two features.

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:125%; text-align:left;"><strong>üëâ Finding the Features Having the Highest Correlation:</strong></p>


**The following code snippet finds the ``two features`` in a DataFrame that have the ``highest correlation with each other``.**

````python
# Initialize variables to track the maximum correlation and corresponding features
max_corr = 0
max_corr_features = []

# Iterate through the correlation matrix to find the highest correlation pair
for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):
        corr_value = correlation_matrix.iloc[i, j]
        
        # Update max_corr and max_corr_features if a higher correlation is found
        if corr_value > max_corr:
            max_corr = corr_value
            max_corr_features = [correlation_matrix.columns[i], correlation_matrix.columns[j]]

# Output the features with the highest correlation
print(f"The two features with the highest correlation: {max_corr_features[0]} & {max_corr_features[1]}")
print(f"Correlation value: {max_corr}")
````

This code effectively identifies and outputs the **``two features with the highest correlation``** in our dataset, along with their correlation value. It's a valuable step in feature selection and understanding relationships within our data.

The **``warning message``** above is a **``"SettingWithCopyWarning"``** in Pandas. This warning occurs when we are **``trying to modify (assign a value) a slice of a DataFrame that is a copy of the original data``**, rather than a view of it, which can lead to unexpected behavior.

    - df2.loc['Afghanistan', 'ladder_score'] = mean_ladder_by_regional["South Asia"]

By using this approach, we directly update the DataFrame without creating a copy, which helps avoid the warning.

**The following User Defined Function ``long_life(life)`` ``classifies life expectancy values`` into categories based on the ``25th percentile``, ``mean``, and ``75th percentile``.**

````python
def long_life(life):
    if life >= _75 :
        return 'Long Life expectancy'
    elif life > mean  and life < _75:
        return 'Long life expectancy'
    elif life  > _25  and life  <= mean:
        return 'Normal life expectancy'
    else:
        return 'Low life expectancy'
````

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:110%; text-align:left;"><strong>-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</strong></p>


**To normalize the investment_risk_factor values using min-max scaling, we can define a function that applies the normalization to our data.**
- This will help visualize the differences between regions more effectively, especially when the values are tightly packed.

````python
# Normalization using min-max scaling

# Define the Min-Max Normalization Function:
def min_max_normalize(data):
    min_val = data.min()
    max_val = data.max()
    return (data - min_val) / (max_val - min_val)
````




<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:110%; text-align:left;"><strong>-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</strong></p>

### BONUS User Defined Functions (DEFs) for detecting potential outliers:

- #### <p style="background-color:#FFFFFF; font-family:newtimeroman; color:blue; font-size:125%; text-align:left;"><strong>ü§îüí≠ How to detect outliers applying Tukey's Fence Rule‚ùì</strong></p>


<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:110%; text-align:left;"><strong>-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</strong></p>

**The following User Defined Function ``find_outliers`` is a useful utility for detecting outliers in a specified column of a DataFrame based on the ``Interquartile Range (IQR)``.**

````python
def find_outliers(dataframe, column_name, threshold=1):
    """
    Identify outliers in a specified column of a DataFrame using the Interquartile Range (IQR) method.

    Parameters:
    ----------
    dataframe : pandas.DataFrame
        The DataFrame containing the data.
    column_name : str
        The name of the column in which to identify outliers.
    threshold : float, optional
        The multiplier for the IQR to define outliers. 
        Default is 1, which is a less common threshold. 
        A typical threshold is 1.5 for standard outlier detection.

    Returns:
    -------
    tuple : (pandas.DataFrame, pandas.DataFrame)
        - A DataFrame with an additional column "Outlier_Column" marking 
          each entry as "Not Outlier" or the name of the column if it is an outlier.
        - A DataFrame containing only the rows identified as outliers.
    """
    
    # Create a copy of the dataframe
    dataframe_copy = dataframe.copy()
    
    # Get the values of the specified column
    column = dataframe_copy[column_name]
    
    # Calculate Quartile 1 (Q1) and Quartile 3 (Q3) values
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    
    # Calculate the Interquartile Range (IQR) value
    iqr = q3 - q1
    
    # Calculate the lower and upper thresholds to identify outliers
    lower_threshold = q1 - threshold * iqr
    upper_threshold = q3 + threshold * iqr
    
    # Create a mask to identify outliers
    outlier_mask = (column < lower_threshold) | (column > upper_threshold)
    
    # Select the rows with outliers
    outliers = dataframe_copy[outlier_mask]
    
    # Mark the outliers in a new column named "Outlier_Column"
    dataframe_copy.loc[outlier_mask, "Outlier_Column"] = column_name
    
    # Define the NaN values in "Outlier_Column" as "Not Outlier"
    dataframe_copy["Outlier_Column"].fillna("Not Outlier", inplace=True)
    
    return dataframe_copy, outliers
````

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:110%; text-align:left;"><strong>-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</strong></p>

**The following User Defined Function ``find_outliers`` ``identifies outliers`` based on the ``mean`` and ``standard deviation`` of a specified column in a DataFrame.**

````python
def find_outliers(dataframe, column_name, threshold=1):
    """
    Identify outliers in a specified column of a DataFrame using the mean and standard deviation method.

    Parameters:
    ----------
    dataframe : pandas.DataFrame
        The DataFrame containing the data.
    column_name : str
        The name of the column in which to identify outliers.
    threshold : float, optional
        The number of standard deviations from the mean to determine outliers. 
        Default is 1. A typical threshold for outlier detection is often set to 2 or 3.

    Returns:
    -------
    tuple : (pandas.DataFrame, pandas.DataFrame)
        - A DataFrame with an additional column "Outlier_Column" marking 
          each entry as "Not Outlier" or the name of the column if it is an outlier.
        - A DataFrame containing only the rows identified as outliers.
    """
    
    # Create a copy of the DataFrame
    dataframe_copy = dataframe.copy()
    
    # Get the values of the specified column
    column = dataframe_copy[column_name]
    
    # Calculate the mean and standard deviation of the column
    mean = column.mean()
    std = column.std()
    
    # Calculate the lower and upper thresholds to determine outliers
    lower_threshold = mean - threshold * std
    upper_threshold = mean + threshold * std
    
    # Create a mask to identify outliers
    outlier_mask = (column < lower_threshold) | (column > upper_threshold)
    
    # Select the rows with outliers
    outliers = dataframe_copy[outlier_mask]
    
    # Mark the outliers in a new column named "Outlier_Column"
    dataframe_copy.loc[outlier_mask, "Outlier_Column"] = column_name
    
    # Define the NaN values in "Outlier_Column" as "Not Outlier"
    dataframe_copy["Outlier_Column"].fillna("Not Outlier", inplace=True)
    
    return dataframe_copy, outliers
````

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:red; font-size:125%; text-align:left;"><strong>ü§îüí≠ Why do they give different results‚ùì</strong></p>

The **``threshold=2 in Tukey's Fence Rule``** and the **``threshold=2 when using standard deviation``** do **``NOT have the same meaning``**. They use different approaches and can produce different results.

- üëâ When using a **``standard deviation-based approach``**, the **``threshold=2``** indicates that we consider a value as an outlier if it is **``more than 2 standard deviations``** away from the mean. This approach is based on the distribution of the data and defines outliers using the standard deviation.

- üëâ On the other hand, **``Tukey's Fence Rule``** is based on the quartiles of the data. The **``threshold=2``** determines the **``lower``** and **``upper``** threshold values by **``multiplying the IQR (Interquartile Range) value``**. If a value is smaller than Q1 - 2 * IQR or larger than Q3 + 2 * IQR, it is considered as an outlier. This method relies on the central tendency and distribution of the data.

Therefore, while the **``threshold=2``** has a **``different meaning``** in a standard deviation-based approach, it has a different meaning in Tukey's Fence Rule. Both methods are different approaches to identifying outliers based on different statistical and distributional characteristics.

## <p style="background-color:#FFFFFF; font-family:newtimeroman; color:blue; font-size:115%; text-align:left;"><strong>üëâ How to detect Skewness & Kurtosis?</strong></p>


**Skewness and Kurtosis:**

[Skewness](https://www.investopedia.com/terms/k/kurtosis.asp) and [Kurtosis](https://www.investopedia.com/terms/s/skewness.asp) are two important statistical terms used in analyzing the distribution of data. They help us understand how the data is spread around the mean and how much it deviates from a normal distribution.

- **Skewness:**

  - Skewness measures the symmetry of the data distribution.
  - In a normal distribution, the skewness value is 0.
  - If the skewness value is positive, the distribution is right-skewed (with more data points above the mean).
  - If the skewness value is negative, the distribution is left-skewed (with more data points below the mean).

- **Kurtosis:**

  - Kurtosis measures the peakedness or flatness of the data distribution.
  - In a normal distribution, the kurtosis value is 3, which is Mesokurtic distribution used to describe the outlier characteristic of a probability distribution that is close to zero..
  - If the kurtosis value is greater than 3, the distribution is called "leptokurtic" and has heavier tails. This means there are more data points near the mean and more extreme data points.
  - If the kurtosis value is less than 3, the distribution is called "platykurtic" and is flatter than a normal distribution. This means there are fewer data points near the mean and fewer extreme data points.

- **``Q3 (Third Quartile)``**: Indicates that 75% of the perceptions of corruption scores are below this value.

These quartiles are useful for understanding the spread and central tendency of the data, and they can also be used to calculate the interquartile range (IQR), which is IQR=Q3‚àíQ1. The IQR is often used to identify outliers in the data.


<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:110%; text-align:left;"><strong>-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</strong></p>

**The following code ``calculates the Interquartile Range (IQR)`` using the ``first (Q1)`` and ``third quartile (Q3)`` values we obtained earlier.**

- A larger IQR indicates greater variability in the middle half of the data, while a smaller IQR suggests that the data points are closer together.
- The IQR is also used in identifying outliers. Typically, any data point that falls below Q1‚àí1.5√óIQR or above Q3+1.5√óIQR is considered an outlier.

**The following code ``calculates the lower limit`` for identifying outliers using the Interquartile Range (IQR) and the first quartile (Q1).**

- If the **``upper_lim value``** is, for instance, **``1.0745``**, then **``any perception of corruption score above 1.0745 would be considered an outlier``**.
- This information is crucial for understanding the distribution of your data and for identifying values that may need further scrutiny.

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:green; font-size:125%; text-align:left;"><strong>üëâ Let's apply the User Defined Function created above for detecting potential outliers:</strong></p>

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:110%; text-align:left;"><strong>-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</strong></p>


**Previously a User Defined function ``find_outliers`` was defined to find outliers in a DataFrame. Below is an example of how we might implement such a function, along with how to use it with your df_CE_Europe DataFrame.**

````python
def find_outliers(df, column, threshold=1.5):
    # Calculate Q1 and Q3
    Q1 = np.percentile(df[column], 25)
    Q3 = np.percentile(df[column], 75)
    IQR = Q3 - Q1
    
    # Calculate the lower and upper bounds
    lower_lim = Q1 - threshold * IQR
    upper_lim = Q3 + threshold * IQR
    
    # Identify outliers
    outliers = df[(df[column] < lower_lim) | (df[column] > upper_lim)]
    
    # Create a modified DataFrame without the outliers
    modified_df = df[~df.index.isin(outliers.index)]
    
    return modified_df, outliers
````

````python
# Find outliers
modified_df, outliers = find_outliers(df_CE_Europe, 'perceptions_of_corruption', threshold=1.5)

# Print the results
print("Modified DataFrame:")
display(modified_df)  # Assuming you're using a Jupyter notebook, or replace with print(modified_df)
print("\nOutliers:")
display(outliers)  # Assuming you're using a Jupyter notebook, or replace with print(outliers)
````

**Function Definition:**
- The find_outliers function takes a DataFrame (df), a column name (column), and a threshold for outlier detection (default is 1.5).
- It calculates the first (Q1) and third (Q3) quartiles, computes the IQR, and then determines the lower and upper limits for identifying outliers.

**Identifying Outliers:**
- It creates a DataFrame of outliers by filtering for values that fall outside the calculated limits.
- The modified DataFrame is created by excluding the outliers.

**Returning the Results:**
- The function returns both the modified DataFrame (without outliers) and the DataFrame of outliers.

**Printing Results**:
- The display() function is typically used in Jupyter notebooks for better formatting. If we're working in a different environment, we can simply use print(modified_df) and print(outliers) instead.
- The printed results will show us the modified **``DataFrame WITHOUT outliers``** and a **``separate DataFrame containing the identified outliers``**.
