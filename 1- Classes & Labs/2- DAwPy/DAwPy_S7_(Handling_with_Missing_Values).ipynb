**[Working with Missing Data in Pandas](https://pandas.pydata.org/docs/user_guide/missing_data.html)**

- Missing Data can occur when ``no information`` is provided for one or more items or for a whole unit. Missing Data is a very ``big problem in a real-life scenarios``. Pandas represents missing values with ``NaN`` (``Not a Number``). Missing Data can also refer to as ``NA`` (an acronym for ``Not Available``) values in Pandas. In DataFrame sometimes many datasets simply arrive with missing data, either because it exists and was ``not collected`` or it ``never existed``. 

**[When and Why Is Data Missed?](https://www.tutorialspoint.com/python_pandas/python_pandas_missing_data.htm)**

- Let us consider an online survey where different users being surveyed for a product. ``Many a times, people do NOT share all the information`` related to them. For Example, different users being surveyed may choose ``not to share their income``, some users may choose ``not to share the address``. Few people share their experience, but not how long they are using the product; few people share how long they are using the product, their experience but not their contact information. Thus, `in some or the other way a part of data is always missing``, and this is very ``common in real time``.


**[Missing Value Handling Methods](https://becominghuman.ai/imputing-missing-values-f00a770d9cc4)**

- **``Delete all instances with at least a missing value``**: If you do this method then, the size of the dataset should be large enough to represent the population without those instances.<br>
- **``Recover redoing the experiment or contacting the subject``**: This is one of the high-cost methods to recover the values, redoing an experiment takes time and cost money, moreover if it's a public opinion kind of dataset, then acts like GDPR restrict future uses to contact the subject in some cases.<br>
- **``Domain Experts Guessing``**: By consulting the domain experts it is possible to guess the missing values with their experience but this method is a hectic and highly cost method due to the high hourly rates of experts.<br>
- **``Fill with mean``**: This is one of the easiest and most primary methods of imputing missing values, by filling each missing value with the average of the attribute. This is not always good to use because it reduces the variability of the data but in some cases makes sense.<br>
- **``Fill with median``**: Similar to the above method the problem is it reduces the variability of the data. Moreover, if the missing values are distributed then it could increase the noise of the data as well.<br>
- **``Using Machine Learning algorithms``**: There are two different ways of using ML for imputation, first the multiple-regression analysis, by creating a regression equation using the rest of the data it is possible to predict missing values. This method needs more complete instances to train the model. Finally, the most common and popular method is the Multiple Imputation. where there are multiple missing values based on the correlations for the missing data and then averages the simulated datasets by incorporating random errors in your predictions

### <p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:120%; text-align:left;"><strong>Operations with ``NaN``:</strong></p>

The following statements demonstrate various behaviors of ``NaN`` and ``floating-point comparisons`` in Python using NumPy. Understanding these behaviors is crucial for handling missing values and performing numerical operations accurately.

When performing **``mathematical operations with NaN``** (Not a Number) in Python, especially using NumPy or pandas, the results can often be **``counterintuitive``**. 

**``Basic Operations with NaN:``**
- **``Addition``**: Adding NaN to any number or another NaN ``results in NaN``.
- **``Subtraction``**: Subtracting NaN from any number or another NaN also ``results in NaN``.
- **``Multiplication:``** Multiplying NaN by any number or another NaN ``results in NaN``.
- **``Division``**: Dividing any number by NaN or dividing NaN by any number ``results in NaN``.
- **``Exponential``**: Any operation that involves NaN will ``return NaN``.
<hr>

**``Comparisons Involving NaN:``**
- **``Equality``**: ``NaN is NOT equal to NaN``. This is a defining characteristic of NaN.
- **``Inequality``**: ``Any comparison with NaN`` (greater than, less than, etc.) will ``return False``.
<hr>

**``Summary:``**
- **``Propagation of NaN``**: Operations involving ``NaN generally propagate the NaN value``, meaning ``if any operand is NaN, the result will also be NaN``.
- **``Special Comparisons``**: NaN is ``treated uniquely in comparisons``, and it is essential to handle it carefully in data analysis to avoid misleading results.

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:110%; text-align:left;"><strong>-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</strong></p>


```python
# 1. Subtracting NaN from NaN
print(np.nan - np.nan)  # Output: nan (any operation involving NaN yields NaN).  

# 2. Adding NaN to NaN
print(np.nan + np.nan)  # Output: nan (any operation involving NaN yields NaN).

# 3. Adding 10 to NaN
print(np.nan + 10)  # Output: nan (any operation involving NaN yields NaN).

# 4. Multiplying 0 by NaN
print(0 * np.nan)  # Output: nan (any operation with NaN results in NaN).

# 5. Subtracting 10 from NaN
print(np.nan - 10)  # Output: nan (any operation involving NaN yields NaN).

# 6. Checking if NaN is equal to NaN
print(np.nan == np.nan)  # Output: False (NaN is not considered equal to itself).

# 7. Checking if infinity is greater than NaN
print(np.inf > np.nan)  # Output: False (comparison with NaN yields False).

# 8. Checking if NaN is NaN
print(np.nan is np.nan)  # Output: True (both refer to the same NaN object).

# 9. Checking if NaN is in a list containing NaN
print(np.nan in [np.nan])  # Output: True (list contains NaN).

# 10. Checking if NaN is in a set containing NaN
print(np.nan in set([np.nan]))  # Output: True (sets can contain NaN).

# 11. Checking floating-point equality
print(0.3 == 3 * 0.1)  # Output: False (due to floating-point precision issues).
```

- **``Equality``** refers to the concept that most Python programmers know as **``‚Äú==‚Äù``**. This is used to ask Python **``whether the content of the variable is the same as the content of another variable``**.The **``content of NaN``** is **``NEVER equal to the content of another NaN``**.

Python assigns an **``id to each variable``** that is created, and ids are compared when Python looks at the identity of a variable in an operation. However, **``np.nan``** is a **``single object``** that **``always has the same id``**, no matter which variable you assign it to.

- Filling missing values with proper values can be done in several ways, depending on the context of your data and the analysis you want to perform. Here are some common methods to fill missing values in a pandas DataFrame with appropriate values:

  - **``1. Filling with the Mean``**: You can fill missing values in a numerical column with the mean of that column. Good for numerical data when the distribution is not skewed (mean) or is skewed (median).
  - **``2. Filling with the Median``**: Filling missing values with the median can be more robust, especially for skewed distributions. Good for numerical data when the distribution is not skewed (mean) or is skewed (median).
  - **``3. Forward Fill (FFill)``**: This method fills missing values with the last valid observation. Useful in time series data where you want to carry forward the last observation.
  - **``4. Backward Fill (BFill)``**: This method fills missing values with the next valid observation. Useful when the future value is a better estimate for the missing value.

- **``s.interpolate()``**: This method fills NaN values by estimating them based on surrounding values.

- **``s.interpolate()``**: This method fills NaN values by estimating them based on surrounding values.

### üì¢‚ùóüö®üî• **``Important Note``**: When you have ``more than one mode`` in a dataset (i.e., a multimodal distribution), it can complicate the process of filling missing values for categorical variables. Here are some considerations and potential approaches:

- **``Issues with Multiple Modes:``**
  - **``Ambiguity``**: If there are multiple modes, choosing one to fill missing values may NOT represent the data accurately, as each mode may reflect different categories that are equally common.
  - **``Loss of Information``**: By selecting only one mode, you might lose important information about the distribution of categories in your dataset.
- **``Strategies to Handle Multiple Modes:``**
  - **``Random Selection:``**
    - If there are multiple modes, you can ``randomly select one`` of the modes to fill the missing values. This introduces some variability but can be a reasonable approach if all modes are equally valid.

  - **``Fill with All Modes:``**
    - Instead of replacing missing values with a single mode, you could ``create a new category`` like "Other" or "Multiple Modes" to indicate that there were multiple common categories. This preserves the information about the modes.

  - **``Use Frequency Distribution:``**
    - Analyze the frequency of each mode and decide to fill missing values with the mode that has the highest frequency. If all modes have the same frequency, you can choose any one of them or use random selection.
  - **``Consider Domain Knowledge:``**
    - Use your understanding of the data context to decide which mode makes the most sense to use for filling in missing values.
    
  - üìå **``Using Grouping to Fill Missing Values:``** selecting the mode based on grouping by other features is indeed a highly effective strategy for filling missing values in categorical variables. This approach leverages relationships in the data, allowing for a more context-aware imputation.
    - **``Group by Other Features``**: You can group the DataFrame by one or more relevant features and then calculate the mode for each group. This allows you to fill missing values in a way that respects the structure of your data.
    - **``Impute Missing Values``**: Fill the missing values in the categorical variable with the mode of the relevant group.

  - **``Filling Strategy``**: For each group, determine the most frequent category (mode) and use it to fill missing values in the categorical variable. This respects the context of the data, as it uses the most common value relevant to that specific group.

### <p style="background-color:#FFFFFF; font-family:newtimeroman; color:blue; font-size:120%; text-align:left;"><strong>e. Filling with Interpolation:</strong></p>


The phrase **``"Interpolating missing values"``** refers to the **``process of estimating and filling in missing data points in a dataset based on the values surrounding them``**. This is particularly useful in datasets where values are expected to change in a ``continuous manner``, such as ``time series data`` or ``measurements taken at regular intervals``.

Pandas **``dataframe.interpolate()``** function is basically used to fill NA values in the dataframe or series. But, this is a very powerful function to fill the missing values. It uses various interpolation technique to fill the missing values rather than hard-coding the value [Source 01](https://www.geeksforgeeks.org/python-pandas-dataframe-interpolate/), [Source 02](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.interpolate.html) & [Source 03](https://note.nkmk.me/en/python-pandas-interpolate/)

- **``Key Concepts of Interpolation:``**
   - **``Estimation``**: Interpolation uses existing data points to estimate values at missing points. This helps maintain the integrity of the dataset without introducing arbitrary values.
   - **``Methods``**: Various methods can be used for interpolation, including:
   - **``Linear Interpolation``**: Estimates the missing value by connecting two known points with a straight line.
   - **``Polynomial Interpolation``**: Fits a polynomial curve to the known data points to estimate missing values.
   - **``Spline Interpolation``**: Uses piecewise polynomials (splines) to create a smooth curve that passes through known data points.
   - **``Time-based Interpolation``**: In time series data, methods can take into account the time dimension for more accurate estimates.
- **``Applications``**: Interpolation is commonly used in:
   - **``Data Cleaning``**: Filling gaps in datasets before analysis.
   - **``Signal Processing``**: Estimating missing readings in sensor data.
   - **``Economics and Finance``**: Filling in missing economic indicators or stock prices.
