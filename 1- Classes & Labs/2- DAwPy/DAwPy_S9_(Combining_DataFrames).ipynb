Combining DataFrames in pandas allows us to merge or concatenate multiple datasets into a single DataFrame, facilitating data analysis and processing. This can be achieved using functions like **``pd.concat()``** for **``vertical``** or **``horizontal``** concatenation, and **``pd.merge()``** for joining DataFrames based on **``common columns``** or **``indices``**. Properly combining DataFrames helps maintain data integrity and simplifies the analysis workflow by consolidating related information into one structure.

There are 4 **main ways** of combining DataFrames together: **``Appending``**, **``Concatenating``**, **``Joining``** and **``Merging``**. In this lecture we will discuss these 4 methods with examples. [A Detailed Video Source](https://www.youtube.com/watch?v=g7n1MZyYjOM).

 - **Append** is the specific case(axis=0, join='outer') of concat.

 - **Concat** gives the flexibility to join based on the axis( all rows or all columns).
 
 - **Join** is based on the indexes (set by set_index)/key columns on how variable =['left','right','inner','outer'].

 - **Merge** is based on any particular column each of the two dataframes, this columns are variables on like 'left_on', 'right_on', 'on'.

- [**concatenate()**](https://pandas.pydata.org/docs/reference/api/pandas.concat.html) method in Pandas ``combines objects`` along a **particular axis** with optional set logic along the other axes.


**SYNTAX:** <b>``pandas.concat(objs, axis=0, join='outer', ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=True)``</b>

**``Concatenation``** in pandas refers to the **``process of joining two or more DataFrames along a specified axis (rows or columns)``**. This is typically done using the **``pd.concat()``** function, which allows for **``vertical stacking``** (adding rows) or **``horizontal alignment``** (adding columns). Concatenation is useful for combining related datasets, maintaining consistency in data structure, and facilitating streamlined data analysis. It can also include options like resetting the index to ensure a clean and organized output.


- [**concatenate()**](https://pandas.pydata.org/docs/reference/api/pandas.concat.html) method in Pandas **``combines objects``** along a **particular axis** with optional set logic along the other axes.

Concatenation basically ``glues together DataFrames``. Keep in mind that **``dimensions should match``** **``along the axis``** we are concatenating on. ### <p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:115%; text-align:left;"><strong>üì¢‚ùóüö® Understanding how the axis parameter works in Pandas is crucial when concatenating DataFrames:</strong></p>

#### <p style="background-color:#FFFFFF; font-family:newtimeroman; color:#008000; font-size:125%; text-align:left;"><strong>üìå What is axis‚ùì</strong></p>

   - In Pandas, the **``axis``** parameter determines the **``direction``** along which **``operations are performed``**. It can take two values:

     - **``axis=0``**: Refers to the **``vertical direction (rows)``**. When we concatenate along this axis, we are stacking DataFrames on ``top of each other``.
       - When concatenating along axis=0, the DataFrames are appended as new rows. This is useful when you have similar structures and want to combine them into one DataFrame.
     - **``axis=1``**: Refers to the **``horizontal direction (columns)``**. When we concatenate along this ``axis=0``, we are placing DataFrames ``side by side``.
       - When concatenating along **``axis=1``**, the DataFrames are merged as **``new columns``**. This is useful when we want to combine **``different features``** of the **``same observations``**.

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:110%; text-align:left;"><strong>-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</strong></p>


**The following code concatenates ``multiple DataFrames`` (df1, df2, and df3) along the ``default axis``, which is ``axis=0`` (i.e., stacking them vertically).** 

```python
# Default = rows onto rows

# Concatenate DataFrames vertically on axis=0
pd.concat([df1, df2, df3])
```

- Notice that the concatenation is performed **``row-wise``**. Also, the **``indices``** from the source DataFrame objects are **``preserved in the output``**.

  - **``Concatenating``**: The pd.concat() function takes a list of DataFrames and combines them. **``By default``**, it **``stacks them vertically (along rows)``**.
  - **``Vertical Stacking``**: The DataFrames are **``stacked vertically``**, resulting in a ``new DataFrame`` that ``includes``;
     - all rows from ``df1``, 
     - followed by ``all rows from df2``, and then 
     - ``all rows from df3``.
  - **``Index Duplication``**: The **``indices``** **``from the original DataFrames``** are **``preserved``**, which leads to **``duplicated indices``** (0 and 1 appear multiple times).

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:110%; text-align:left;"><strong>-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</strong></p>

**The following code concatenates ``multiple DataFrames`` ``vertically`` while ``resetting the index`` in the resulting DataFrame.** 

```python
# Concatenate DataFrames vertically on axis=0 by resetting the index

pd.concat([df1, df2, df3], ignore_index=True)
```

- **``pd.concat()``**: This function is used to combine DataFrames. It takes a list of DataFrames as input.

- **``Vertical Stacking``**: The DataFrames are ``stacked on top of each other``, creating a ``single DataFrame`` with;
   - all rows from ``df1``, 
   - followed by ``all rows from df2``, and then 
   - ``all rows from df3``.
- **``ignore_index=True``**: This parameter resets the index in the resulting DataFrame to ensure it is continuous, starting from 0.

  - **``New Index``**: The resulting DataFrame has a ``new index`` that runs;
     - ``from 0 to the total number of rows minus one``, 
     - **``eliminating any index duplication``** from the original DataFrames.

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:110%; text-align:left;"><strong>-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</strong></p>

**The following code concatenates ``specific columns`` from the DataFrames ``df1``, ``df2``, and ``all of df3``.**

```python
# Concatenate specific columns from df1, df2, and df3 vertically on axis=0

pd.concat([df1.iloc[:, :2], df2.iloc[:, 1:], df3])
```

**Explanation of the Slices:**
- **``df1.iloc[:, :2]``**: This selects ``all rows of df1`` but ``only the first two columns`` (``A`` and ``B``).
- **``df2.iloc[:, 1:]``**: This selects ``all rows of df2`` but from the ``second column to the end`` (``B``, ``C``, and ``D``).
- **``df3``**: This includes ``all rows`` and ``all columns`` of ``df3``.
<hr>

**Key Points:**
- **Column Alignment:**
   - The resulting DataFrame has columns ``A, B from df1``, then ``B, C, D from df2``, and ``all columns from df3``.
- **Missing Values:**
   - Rows corresponding to ``df1`` and ``df2`` will have ``NaN`` for columns that are not filled from those DataFrames.
- **Index Preservation:**
   - The ``original indices`` of the DataFrames are ``preserved``. You may see ``repeated indices` (0, 1, etc.) from ``df2`` and ``df3``. To reset the index, you can use ``ignore_index=True``.

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:110%; text-align:left;"><strong>-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</strong></p>

**The following code concatenates the DataFrames ``df1``, ``df2``, and ``df3`` ``horizontally`` (side by side).** 

```python
# Concatenate DataFrames horizontally (side by side) on axis=1

pd.concat([df1, df2, df3], axis=1)
```

**Key Points:**
- **``Horizontal Stacking``**: The DataFrames are ``combined side by side``, meaning ``each row`` from df1, df2, and df3 is ``aligned by index`. Note that they ``share the same index``.
- **``Index Preservation``**: The ``original indices`` from each DataFrame are ``preserved``. Since ``all DataFrames`` have the ``same number of rows``, the ``indices align perfectly``.
- **``Column Overlap``**: Since df1, df2, and df3 all have columns named A, B, C, and D, the resulting DataFrame will have ``duplicate column names``. If you want ``to avoid`` this, you ``can rename the columns before concatenation``.

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:110%; text-align:left;"><strong>-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</strong></p>

**The following code concatenates ``specific slices (parts)`` of the ``three DataFrames`` ``horizontally``.** 

```python
# Concatenate specific rows from df1, df2, and df3 horizontally (side by side)

pd.concat([df1.iloc[:2], df2.iloc[1:], df3], axis=1)
```

**Explanation of the Output:**
- **``Row Alignment``**:
   - The ``first two rows`` from ``df1`` are ``present``.
   - The ``second to last rows`` from ``df2`` are ``aligned starting from index 1``.
   - ``All rows`` from`` df3`` are ``included``, starting from index 0.
- **``NaN Values``**:
   - Since the ``indices do NOT fully overlap`` (e.g., there are no values from df2 for the first row of df1), ``NaN`` values ``appear where data is missing``.

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:110%; text-align:left;"><strong>-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</strong></p>

## <p style="background-color:#E51A59; font-family:newtimeroman; color:#FFF9ED; font-size:175%; text-align:center; border-radius:10px 10px;">Joining</p>

**SYNTAX:** <b>``DataFrame.join(other, on=None, how='left', lsuffix='', rsuffix='', sort=False)``</b>

Joining is a convenient method for combining the columns of two potentially ``differently-indexed DataFrames`` into a single result DataFrame.

It will combine all the columns from the two tables, with the ``common columns renamed with the defined lsuffix and rsuffix``.

The way that rows from the two tables are combined is defined by how **üßê [Source 01](https://www.educative.io/edpresso/three-ways-to-combine-dataframes-in-pandas), [Source 02](https://www.educba.com/python-pandas-join/), [Source 03](https://studymachinelearning.com/difference-between-merge-join-and-concatenate/), [Source 04](https://www.educba.com/pandas-dataframe-join/) & [Source 05](https://morioh.com/p/cc5cec94fb2d)**.

 -  The join method in Pandas is used to combine DataFrames based **``on their index``**. It is a convenient method when we want to join DataFrames with a **common index**.
 -  **``By default``**, join performs a **``left join``**, which means it keeps all the rows from the left DataFrame and only adds the matching rows from the right DataFrame. Other types of joins (inner, outer, right) can be specified using the **``how parameter``**.
 -  Joining DataFrames using join is particularly useful when we have DataFrames with **``different columns but share a common index``**, and we want to bring their data together based on that **``common index``**.

- **``Merging Implications``**: If a join operation (e.g., an outer join) is performed using the indices:
  - The resulting DataFrame would include all unique indices from both DataFrames: K0, K1, K2, and K3.
  - For indices K0 and K2, the corresponding values from both DataFrames would be merged. However, for K1, there would be NaN values for columns C and D, and for K3, there would be NaN values for columns A and B.

The ``on`` parameter works in the command ``left8.join(right8.set_index('key'), on='key')`` because of the way pandas handles the join operation when ``one DataFrame is using a column for the join`` and ``the other DataFrame is using an index``.

### <p style="background-color:#FFFFFF; font-family:newtimeroman; color:#008000; font-size:125%; text-align:left;"><strong>üìåü§ûüí° WRAP UP:</strong></p>


In Pandas, concatenate, join, and merge are three distinct methods for combining DataFrames, each suited for specific scenarios. Concatenation is used when we want to stack multiple DataFrames either vertically or horizontally. 
- For instance, imagine a school compiling student grades from different terms. If each term‚Äôs grades are stored in separate DataFrames, we can concatenate them to create a comprehensive report card that includes all terms' grades in one sheet. This is done using pd.concat().

Joining is ideal when we want to combine DataFrames based on their indices. 
- For example, consider two departments in a company: the HR department maintains employee IDs as indices, while the Payroll department tracks salaries using these same IDs. By using the .join() method, you can easily align and combine employee information with their respective salaries based on these indices, creating a unified view of employee data.

On the other hand, merging is most powerful when we need to combine DataFrames based on specific (common) key columns rather than indices. 
- For instance, if we have a DataFrame containing customer orders and another DataFrame containing customer details, we can merge them on a common key, such as customer ID. This allows us to create a detailed report that links each order with the corresponding customer information, utilizing the pd.merge() function.

<p style="background-color:#FFFFFF; font-family:newtimeroman; color:#FF0000; font-size:125%; text-align:left;"><strong>‚ö†Ô∏è Be Careful About Duplicates When Using Merge ‚ùó‚ùó‚ùó</strong></p>


- When using the merge function in Pandas, it's important to be cautious, as it can lead to duplicate rows in the resulting DataFrame if there are multiple matching keys in either of the DataFrames being combined. This is especially relevant when merging on non-unique keys, as each combination of matched rows will generate a Cartesian product, potentially inflating the size of the DataFrame and complicating subsequent analyses. Always check for duplicates in your key columns before merging to ensure the integrity of your data.

In summary, choose concatenation for stacking DataFrames when they share similar structures, use joining for aligning data based on indices, and opt for merging when you need to combine DataFrames based on specific key columns to gain comprehensive insights.

#### <p style="background-color:#FFFFFF; font-family:newtimeroman; color:#0000FF; font-size:125%; text-align:left;"><strong>üëâ Difference between Merge, join, and concatenate:</strong></p>


The merge() and join() methods are the DataFrame method, not a series method. The concat() method is the pandas‚Äô method which provides the functionality to combine the pandas‚Äô objects such as DataFrame and Series.

**Merge** 

 - The merge() function used to merge the DataFrames with database-style join such as inner join, outer join, left join, right join.
 - Combining exactly two DataFrames.
 - The join is done on columns or indexes.
 - If joining columns on columns, the DataFrame indexes will be ignored.
 - If joining indexes on indexes or indexes on a column, the index will be passed on.
 

**Join** 

 - The join() function used to join two or more pandas DataFrames/Series horizontally.
 - Join() uses merge internally for the index-on-index (by default) and column(s)-on-index join.
 - Aligns the calling DataFrame‚Äôs column(s) or index with the other objects‚Äô index (and not the columns).
 - Defaults to left join with options for right, inner and outer join
 

**Concat** 

 - concatenate two or more pandas DataFrames/Series vertically or horizontally.
 - Aligns only on the index by specifying the axis parameter.
 - Defaults to outer join with the option for inner join
